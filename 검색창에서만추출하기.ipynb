```python
#  '동기부여', '혁신 전략', '시장 변화', '기여', '비전'
storage = []
target = ''
iniSearchKeywords = ['호텔 컨시어지']
iniSearchKeywords = [target + ' ' + word for word in iniSearchKeywords ]
searchKeywords = []
nextSearchKeywords = []
maxDepth = 3

driver = webdriver.Chrome() # 윈도우열기
try :
    driver.get(createDaumURL('지성')) # 페이지열기

    for depth in range(maxDepth) :    
        if debug : print(f'now {depth}th iteration...',)        
        # 검색어 셋팅
        if depth == 0 :
            searchKeywords = iniSearchKeywords.copy()
        else :
            searchKeywords = list(set(nextSearchKeywords.copy()))
            nextSearchKeywords = []
        if debug : print('all the keywords with this depth is below :')
        if debug : print(searchKeywords)

        for searchKeyword in searchKeywords :
            # 검색창선택
            elements = driver.find_elements(By.CSS_SELECTOR, daumClickElementCSSQuery)
            searchBox = elements[0]

            # 검색창 비우기
            searchBox.clear()
            
            # 검색창 입력
            driver.execute_script(f"arguments[0].value='{searchKeyword}';", searchBox)
            searchBox.click()
            time.sleep(1)
               

            # 글자 추출 
            elements = driver.find_elements(By.CSS_SELECTOR, daumDataCSSQuery)
            
            result = [element.text for element in elements]
            
            # 글자 저장
            storage += result 

            # 다음 검색 키워드 준비
            nextSearchKeywords += result
            searchBox.clear()
finally :
    driver.quit()


storage


```
